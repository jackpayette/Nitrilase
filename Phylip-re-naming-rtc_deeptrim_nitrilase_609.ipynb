```python
aln_name   = 'rtc_deeptrim_nitrilase_609.fa'
fasta      = open(aln_name).read()
taxa_names = re.findall('^>(\S+?)[\s|]', fasta, re.M)
for count, name in enumerate(taxa_names):
    fasta = re.sub('^>%s' % name, '>taxa_%i_ ' % count, fasta, flags=re.M)
```

```python
out = open('rtc_deeptrim_nitrilase_609.fas', 'w')
out.write(fasta)
out.close()
```

```python
SeqIO.convert('rtc_deeptrim_nitrilase_609.fas', 'fasta', 'rtc_deeptrim_nitrilase_609.phy', 'phylip')
phylip = open('rtc_deeptrim_nitrilase_609.phy').read()
phylip_key=list()
phylip_dict = {}
phylip_dict2 = {}
for count, name in enumerate(taxa_names):
    phylip = re.sub('^taxa_%i_\s+' % count, '%s  ' % name, phylip, flags=re.M)
    print('tax'+str(count)+'_,'+str(name))
    phylip_key.append(name+'_'+str(count))
    phylip_dict['tax'+str(count)+'_']=name
    phylip_dict2['taxa_'+str(count)+'_']=name
    #print(name+'_'+str(count))
    #print(taxa_names)
```
